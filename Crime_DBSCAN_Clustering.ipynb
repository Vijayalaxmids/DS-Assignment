```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('crime_data.csv')
```

```python
df = df.rename(columns={'Unnamed: 0':'City'})
```

```python
df1 = df.iloc[0:,1:]
df1.head()
```

```python
df1.describe()
```

```python
plt.rcParams['figure.figsize'] = (8,4)
plt.rcParams['figure.dpi'] = 150
```

```python
sns.pairplot(df1)
```

```python
sc = StandardScaler()
```

```python
x = sc.fit_transform(df1)
x
```

```python
db = DBSCAN()
y = db.fit_predict(x)
y
```

```python
from sklearn.metrics import silhouette_score
from sklearn.neighbors import NearestNeighbors
```

```python
neigh = NearestNeighbors(n_neighbors =12).fit(x)
dist, ind = neigh.kneighbors(x)
```

```python
pd.DataFrame(dist)
```

```python
distance = np.sort(dist[:,1], axis = 0)
```

```python
distance
```

```python
sns.set_style('darkgrid')
plt.plot(distance)
```

```python
db = DBSCAN(eps = 1.25, min_samples =3)
y = db.fit_predict(x)
y
```

```python
silhouette_score(x,y)
```

```python
df['cluster'] = y
```

```python
df
```

```python
df[df['cluster'] == -1]
```

```python
sns.pairplot(df, hue = 'cluster',palette='rainbow')
```
